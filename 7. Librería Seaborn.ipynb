```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
sns.histplot(data=tips, x="total_bill", bins=20, kde=True, color="blue")
plt.title("Distribuci√≥n del Total de la Cuenta")
plt.xlabel("Total de la Cuenta ($)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", size="size", palette="coolwarm", alpha=0.8)
plt.title("Relaci√≥n entre Total y Propina por G√©nero")
plt.xlabel("Total de la Cuenta ($)")
plt.ylabel("Propina ($)")
plt.show()
```

```python
numeric_tips = tips.select_dtypes(include=["float64", "int64"])
corr = numeric_tips.corr()
sns.heatmap(corr, annot=True, cmap="YlGnBu", linewidths=0.5, fmt=".2f")
plt.title("Mapa de Calor de Correlaciones")
plt.show()
```

```python
sns.pairplot(data=tips, hue="time", diag_kind="kde", palette="husl")
plt.show()
```

```python
diamonds = sns.load_dataset("diamonds")
sns.boxenplot(data=diamonds, x="cut", y="price", hue="cut", palette="coolwarm")
plt.title("Distribuci√≥n de Precios de Diamantes por Calidad del Corte")
plt.xlabel("Calidad del Corte")
plt.ylabel("Precio ($)")
plt.show()
```

```python
sns.ecdfplot(data=diamonds, x="price", hue="cut", palette="Set2")
plt.title("Distribuci√≥n Acumulativa de Precios de Diamantes")
plt.xlabel("Precio ($)")
plt.ylabel("Proporci√≥n Acumulativa")
plt.show()
```

```python
sns.jointplot(data=diamonds, x="carat", y="price", kind="hex", color="blue")
plt.suptitle("Relaci√≥n entre Peso (Carat) y Precio de Diamantes", y=1.02)
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="violin", hue="sex", split=True, palette="husl")
plt.title("Distribuci√≥n de Total de Cuenta por D√≠a y G√©nero")
plt.show()
```

```python
sns.boxplot(data=diamonds, x="cut", y="price", hue="cut", palette="pastel")
sns.despine(left=True)
plt.title("Precios de Diamantes por Corte (Sin Ejes Laterales)")
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
sns.clustermap(flights_pivot, cmap="coolwarm", annot=True, fmt="d", linewidths=0.5)
plt.title("Mapa de Calor con Agrupamiento de Pasajeros")
plt.show()
```
