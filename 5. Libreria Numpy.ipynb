```python
k = np.array([1, 2, 3])
l = np.array([4, 5, 6])
print(k + l)
```

```python
data = np.array([50, 80, 100, 120])
normalized_data = (data - data.min()) / (data.max() - data.min())
print(normalized_data)
```

```python
mean = data.mean()
std = data.std()
scaled_data = (data - mean) / std
print(scaled_data)
```

```python
A = np.array([[2, 1], [1, 3]])
b = np.array([8, 13])
solución = np.linalg.solve(A, b)
print(solución)
```
