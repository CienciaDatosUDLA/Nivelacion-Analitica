```python
# prompt: Proporciona el código para calcular y mostrar la media, mediana, moda, desviación estándar y percentiles para una columna específica de un DataFrame.​

import pandas as pd

# Suponiendo que el DataFrame se llama 'iris_df' y quieres analizar la columna 'sepal_length'
# Reemplaza 'sepal_length' con el nombre de la columna que te interesa

def analyze_column(df, column_name):
  """Calcula estadísticas descriptivas para una columna dada de un DataFrame."""

  if column_name not in df.columns:
    print(f"Error: La columna '{column_name}' no existe en el DataFrame.")
    return

  # Calcular la media
  mean_value = df[column_name].mean()

  # Calcular la mediana
  median_value = df[column_name].median()

  # Calcular la moda
  try:
    mode_value = df[column_name].mode().iloc[0]  # Toma el primer valor de la moda si hay varios
  except:
      mode_value = "No hay moda única"

  # Calcular la desviación estándar
  std_dev = df[column_name].std()

  # Calcular percentiles (ejemplos del 25%, 50% y 75%)
  percentiles = df[column_name].quantile([0.25, 0.5, 0.75])

  # Mostrar los resultados
  print(f"Análisis de la columna '{column_name}':")
  print(f"Media: {mean_value}")
  print(f"Mediana: {median_value}")
  print(f"Moda: {mode_value}")
  print(f"Desviación estándar: {std_dev}")
  print(f"Percentiles:\n{percentiles}")

# Ejemplo de uso con la columna 'sepal_length' del DataFrame 'iris_df'
analyze_column(iris_df, 'sepal_length')
```

```python
# prompt: Genera gráficos exploratorios para analizar la distribución y las relaciones entre variables en un DataFrame. Incluye histogramas, gráficos de dispersión y un mapa de calor de correlación.​

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ... (tu código anterior) ...

# Histogramas para cada variable numérica
iris_df.hist(figsize=(10, 8))
plt.suptitle("Histogramas de las variables numéricas", fontsize=16)
plt.show()

# Gráficos de dispersión para pares de variables
sns.pairplot(iris_df, hue="species")
plt.suptitle("Gráficos de dispersión para pares de variables", fontsize=16)
plt.show()

# Mapa de calor de correlación
correlation_matrix = iris_df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de calor de correlación")
plt.show()
```

```python
# prompt: Genera gráficos avanzados en seaborn para visualizar tendencias y relaciones en un conjunto de datos. Incluye un pairplot y un gráfico de barras agrupadas.

# Gráfico de barras agrupadas
plt.figure(figsize=(10, 6))
sns.barplot(x='species', y='sepal_length', data=iris_df, ci=None)  # ci=None elimina las barras de error
plt.title('Longitud del sépalo por especie')
plt.xlabel('Especie')
plt.ylabel('Longitud del sépalo (cm)')
plt.show()


# Pairplot con opciones de personalización
sns.pairplot(iris_df, hue="species", diag_kind="kde", markers=["o", "s", "D"], palette="husl") # diag_kind="kde" para la distribución en la diagonal
plt.suptitle("Pairplot con opciones de personalización", fontsize=16)
plt.show()

# Ejemplo de jointplot
sns.jointplot(x='sepal_length', y='sepal_width', data=iris_df, kind='kde', color='skyblue')
plt.show()

# Ejemplo de boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x='species', y='petal_length', data=iris_df, palette="Set3")
plt.title("Boxplot de Petal Length por especie")
plt.show()


# Ejemplo de violinplot
plt.figure(figsize=(8,6))
sns.violinplot(x='species', y='petal_width', data=iris_df, palette="Set2")
plt.title("Violinplot de Petal Width por especie")
plt.show()
```
