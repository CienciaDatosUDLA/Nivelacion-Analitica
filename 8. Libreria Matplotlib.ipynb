```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('titanic')
print(df.head())
```

```python
df.groupby('sex')['survived'].value_counts().unstack().plot(kind='bar', stacked=True)
```

```python
plt.scatter(df['age'], df['fare'], alpha=0.5)
plt.title('Relaci√≥n entre Edad y Precio del Billete')
plt.xlabel('Edad')
plt.ylabel('Precio del billete')
plt.show()
```

```python
df_survived = df[df['survived'] == 1]['age']
df_not_survived = df[df['survived'] == 0]['age']
plt.boxplot([df_survived.dropna(), df_not_survived.dropna()], labels=['Sobrevivi√≥', 'No sobrevivi√≥'])
plt.title('Distribuci√≥n de Edad por Supervivencia')
plt.xlabel('Supervivencia')
plt.ylabel('Edad')
plt.show()
```

```python
plt.hist(df['age'].dropna(), bins=20, alpha=0.7, color='blue', edgecolor='black')
plt.title('Distribuci√≥n de Edad de los Pasajeros')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```
